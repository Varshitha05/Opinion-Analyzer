```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import torch

# Load the dataset
file_path = '/content/longterm_cleaned_comments.csv'  # Update this path
data = pd.read_csv(file_path)
data['Cleaned_Comments'] = data['Cleaned_Comments'].fillna('')
data = data[data['Cleaned_Comments'].apply(lambda x: len(str(x).split()) >= 5)]
# Your themes and keywords
theme_keywords = {
    'ethical' : [
    "morality", "ethics", "principles", "conscience", "dignity",
    "sanctity of life", "justice", "values", "compassion", "choice",
    "autonomy", "responsibility", "integrity", "harm", "benefit",
    "duty", "virtue", "consent", "welfare", "suffering",
    "well-being", "personhood", "moral values", "moral principles", "ethical standards",
    "human rights", "respect for life", "ethical dilemma", "moral obligation", "consequences"
],
    'political' : [
    "legislation", "policy", "government", "regulation", "rights",
    "law", "court", "federal", "state", "political parties",
    "advocacy", "democracy", "governance", "public policy", "civil liberties",
    "judicial", "constitution", "electoral", "campaign", "partisanship",
    "bipartisan", "political action", "political debate", "political ideology", "political discourse",
    "political rights", "policy-making", "issue-framing", "voting", "legislative process"
],
    'social' : [
    "community", "society", "public opinion", "cultural norms", "social change",
    "equality", "activism", "social movement", "awareness", "public health",
    "social impact", "social values", "social responsibility", "welfare", "social norms",
    "discrimination", "inequality", "empowerment", "social media", "community support",
    "public discourse", "social ethics", "cultural impact", "social awareness", "gender roles",
    "family planning", "education", "social welfare", "social rights", "reproductive health"
],
    'legal' : [
    "Supreme Court", "constitutional", "legislation", "Roe v. Wade", "rights",
    "federal", "state law", "judicial", "court decision", "legal challenge",
    "civil liberties", "statute", "regulation", "legal system", "justice",
    "appeal", "litigation", "policy", "lawmaker", "bill",
    "legal rights", "judiciary", "ruling", "compliance", "precedent",
    "legal action", "court ruling", "legal status", "amendment", "legislature"
]
}

# Initialize the SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate theme descriptions by concatenating the keywords for each theme
theme_descriptions = {theme: ' '.join(keywords) for theme, keywords in theme_keywords.items()}

# Generate embeddings for each theme description
theme_embeddings = {theme: model.encode(description, convert_to_tensor=True) for theme, description in theme_descriptions.items()}

# Function to find the closest theme for a comment
def find_theme(comment, theme_embeddings):
    comment_embedding = model.encode(comment, convert_to_tensor=True)
    highest_similarity = -1
    best_theme = None
    for theme, embedding in theme_embeddings.items():
        similarity = util.pytorch_cos_sim(comment_embedding, embedding)[0][0].item()
        if similarity > highest_similarity:
            highest_similarity = similarity
            best_theme = theme
    return best_theme

# Apply the function to each comment in the dataset
data['Theme'] = data['Cleaned_Comments'].apply(lambda comment: find_theme(comment, theme_embeddings))

# Save the results to a new CSV file
data.to_csv('longterm_categorized_comments.csv', index=False)
```
